```python
def date_sorter():
    regex1 = df.str.extract(r'(\d{2})\s(Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)[a-z]*?\s(\d{2,4})')
    regex2 = df.str.extract(r'(Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)[a-z.]*?\s(\d{0,2}),?\s(\d{2,4})')
    regex3 = df.str.extract(r'()(Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)[a-z]*?\s(\d{2,4})')
    regex4 = df.str.extract(r'(\d{1,2})[/-](\d{1,2})[/-](\d{2,4})')
    regex5 = df.str.extract(r'()(\d{1,2})/(\d{2,4})')
    regex6 = df.str.extract(r'()()(\d{4})')
    df_all = regex1.fillna(regex2).fillna(regex3).fillna(regex4).fillna(regex5).fillna(regex6)
    df_all[3] = df_all[0].map(str) + '-' + df_all[1].map(str) + '-' + df_all[2].map(str)
    date = pd.to_datetime(df_all[3])
    
    year_cutoff = pd.to_datetime("2020-01-01")
    date.loc[date > year_cutoff] = date - pd.offsets.DateOffset(years=100)
    
    return pd.Series(date.sort_values().index) #pd.Series(dateform.sort_values().index)
```
